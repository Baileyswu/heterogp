```python
import math
import numpy as np
import pylab as plt
import pandas as pd
def draw(X, ys):
    plt.figure(figsize=(6, 6))
    plt.ylabel("x2")
    plt.xlabel("x1")
    plt.scatter(
        X[:,0].reshape(-1,1),
        X[:,1].reshape(-1,1),
        cmap="coolwarm", 
        c=ys.reshape(-1,1))
    plt.show()

    table = pd.DataFrame(ys)
    table.describe()
    table[0].hist()
def sigma_func(x):
    return math.pow(0.17*math.sin(math.pi*x) + 0.005*x*x,2)
def artificial_center(N, ct, label):
    X1 = np.linspace(-2.5,2.5,N)[:,None]
    X2 = np.random.rand(N,1)
    Y = np.random.rand(N,1)
    for i in range(N):
        sigma = sigma_func(X1[i])
        X2[i] = np.random.normal(ct, sigma, 1)
        Y[i] = label
    X = np.hstack((X1, X2))
    return X, Y

N = 300
X_neg, Y_neg = artificial_center(N, -0.05, -1)
X_pos, Y_pos = artificial_center(N, 0.05, 1)
X = np.vstack((X_neg, X_pos))
Y = np.vstack((Y_neg, Y_pos))

draw(X, Y)
```
